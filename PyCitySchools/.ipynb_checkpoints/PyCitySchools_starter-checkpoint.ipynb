```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
SchoolData_csv = "Resources/schools_complete.csv"
StudentData_csv = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
SchoolData = pd.read_csv(SchoolData_csv)
StudentData = pd.read_csv(StudentData_csv)

# Combine the data into a single dataset.  
SchoolComplete = pd.merge(StudentData, SchoolData, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
SchoolCount = len(SchoolData['School ID'])
StudentCount = len(StudentData['Student ID'])
Budget = (SchoolData['budget'].sum())
AverageMath = SchoolComplete['math_score'].mean()
AverageRead = SchoolComplete['reading_score'].mean()
```

```python
PassingMath = SchoolComplete[(SchoolComplete['math_score'] >= 70)].count()['student_name']
MathPercent = PassingMath / float(StudentCount)*100
PassingRead = SchoolComplete[(SchoolComplete['reading_score'] >= 70)].count()['student_name']
ReadPercent = PassingRead / float(StudentCount)*100
Overall = SchoolComplete[(SchoolComplete['reading_score'] >= 70) & (SchoolComplete['math_score'] >= 70)].count()['student_name']
OverallPercent = Overall / float(StudentCount)*100
```

```python
ADataFrame = pd.DataFrame({'Total Schools': [SchoolCount], 'Total Students': [StudentCount], 
              'Total Budget': [Budget], 'Average Math Score': [AverageMath], 'Average Reading Score': [AverageRead],
             '% Passing Math': [MathPercent], '% Passing Reading': [ReadPercent], '% Overall Passing': [OverallPercent]})
ADataFrame['Total Students'] = ADataFrame['Total Students'].map('{:,}'.format)
ADataFrame['Total Budget'] = ADataFrame['Total Budget'].map('${:,.2f}'.format)


ADataFrame
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
Types = SchoolData.set_index(['school_name'])['type']
Count = SchoolComplete['school_name'].value_counts()

PerSchool = SchoolComplete.groupby(['school_name']).mean()['budget']
PerStudent = PerSchool / Count 

PerSchoolMath = SchoolComplete.groupby(['school_name']).mean()['math_score']
PerSchoolRead = SchoolComplete.groupby(['school_name']).mean()['reading_score']
```

```python
SchoolPassingMath = SchoolComplete[(SchoolComplete['math_score'] >= 70)]

SchoolPassingRead = SchoolComplete[(SchoolComplete['reading_score'] >= 70)]

SchoolOverall = SchoolComplete[(SchoolComplete['reading_score'] >= 70) & (SchoolComplete['math_score'] >= 70)]
```

```python
SPM = SchoolPassingMath.groupby(['school_name']).count()['student_name']/Count * 100
SPR = SchoolPassingRead.groupby(['school_name']).count()['student_name']/Count * 100
SO = SchoolOverall.groupby(['school_name']).count()['student_name']/Count * 100
```

```python
BDataFrame = pd.DataFrame({'School Type': Types, 
                           'Total Students': Count, 
                           'Total School Budget': PerSchool, 
                           'Per Student Budget': PerStudent,
                           'Average Math Score': PerSchoolMath, 
                           'Average Reading Score': PerSchoolRead,
                           '% Passing Math': SPM, 
                           '% Passing Reading': SPR, 
                           '% Overall Passing': SO})


BDataFrame['Total School Budget'] = BDataFrame['Total School Budget'].map('${:,.2f}'.format)
BDataFrame['Per Student Budget'] = BDataFrame['Per Student Budget'].map('${:,.2f}'.format)

BDataFrame
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
Top = BDataFrame.sort_values(['% Overall Passing'], ascending=False)
Top.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth = SchoolComplete[(SchoolComplete['grade'] == '9th')]
tenth = SchoolComplete[(SchoolComplete['grade'] == '10th')]
eleventh = SchoolComplete[(SchoolComplete['grade'] == '11th')]
twelfth = SchoolComplete[(SchoolComplete['grade'] == '12th')]

ninthscores = ninth.groupby(['school_name']).mean()['math_score']
tenthscores = tenth.groupby(['school_name']).mean()['math_score']
eleventhscores = eleventh.groupby(['school_name']).mean()['math_score']
twelfthscores = twelfth.groupby(['school_name']).mean()['math_score']

ByGrade = pd.DataFrame({'9th': ninthscores,
                       '10th': tenthscores,
                       '11th': eleventhscores,
                       '12th': twelfthscores})
ByGrade['9th'] = ByGrade['9th'].map('{:.2f}'.format)
ByGrade['10th'] = ByGrade['10th'].map('{:.2f}'.format)
ByGrade['11th'] = ByGrade['11th'].map('{:.2f}'.format)
ByGrade['12th'] = ByGrade['12th'].map('{:.2f}'.format)

ByGrade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
Rninthscores = ninth.groupby(['school_name']).mean()['reading_score']
Rtenthscores = tenth.groupby(['school_name']).mean()['reading_score']
Releventhscores = eleventh.groupby(['school_name']).mean()['reading_score']
Rtwelfthscores = twelfth.groupby(['school_name']).mean()['reading_score']

ReadByGrade = pd.DataFrame({'9th': Rninthscores,
                       '10th': Rtenthscores,
                       '11th': Releventhscores,
                       '12th': Rtwelfthscores})
ReadByGrade['9th'] = ReadByGrade['9th'].map('{:.2f}'.format)
ReadByGrade['10th'] = ReadByGrade['10th'].map('{:.2f}'.format)
ReadByGrade['11th'] = ReadByGrade['11th'].map('{:.2f}'.format)
ReadByGrade['12th'] = ReadByGrade['12th'].map('{:.2f}'.format)

ReadByGrade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
BinFrame = BDataFrame
Bins = [0, 584.99, 629.99, 644.99, 680]
Groups = ['<$585', '$585-630','$630-645', '$645-680']
BDataFrame['Spending Ranges (Per Student)'] = pd.cut(PerStudent, Bins, labels=Groups, include_lowest=True)
BinFrame
```

```python
SpendingAveMath = BinFrame.groupby(['Spending Ranges (Per Student)']).mean()['Average Math Score']
SpendingAveRead = BinFrame.groupby(['Spending Ranges (Per Student)']).mean()['Average Reading Score']

SpendingPerMath = BinFrame.groupby(['Spending Ranges (Per Student)']).mean()['% Passing Math']
SpendingPerRead = BinFrame.groupby(['Spending Ranges (Per Student)']).mean()['% Passing Reading']
SpendingOverall = BinFrame.groupby(['Spending Ranges (Per Student)']).mean()['% Overall Passing']
```

```python
SpendingFrame = pd.DataFrame({'Average Math Score': SpendingAveMath,
                             'Average Reading Score': SpendingAveRead,
                             '% Passing Math': SpendingPerMath,
                             '% Passing Reading': SpendingPerRead,
                             '% Overall Passing': SpendingOverall})

SpendingFrame['Average Math Score'] = SpendingFrame['Average Math Score'].map('{:.2f}'.format)
SpendingFrame['Average Reading Score'] = SpendingFrame['Average Reading Score'].map('{:.2f}'.format)
SpendingFrame['% Passing Math'] = SpendingFrame['% Passing Math'].map('{:.2f}'.format)
SpendingFrame['% Passing Reading'] = SpendingFrame['% Passing Reading'].map('{:.2f}'.format)
SpendingFrame['% Overall Passing'] = SpendingFrame['% Overall Passing'].map('{:.2f}'.format)

SpendingFrame
```
